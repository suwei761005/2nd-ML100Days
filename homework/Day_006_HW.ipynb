* 未知值，隨意填補 (約定俗成的代入)，如年齡常見 0,999

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
dir_data = './data'
f_app = os.path.join(dir_data, 'application_train.csv')
app_train = pd.read_csv(f_app)
# 先篩選數值型的欄位

dtype_select = ['int','float']

numeric_columns = list(app_train.columns[list(app_train.dtypes.isin(dtype_select))])

# 再把只有 2 值 (通常是 0,1) 的欄位去掉
numeric_columns = list(app_train[numeric_columns].columns[list(app_train[numeric_columns].apply(lambda x:len(x.unique())!=2 ))])
# print("Numbers of remain columns" % len(numeric_columns))
# print(len(numeric_columns))
# 檢視這些欄位的數值範圍
for col in numeric_columns:
#     print(app_train[col])
   temp_train = app_train[col].dropna(axis = 0,how = 'any')
   plt.boxplot(temp_train,
            notch=True,  # notch shape
            sym='bs',     # blue squares for outliers
            vert=True,   # vertical box aligmnent
            patch_artist=True)   # fill with color 
   plt.title(col) 
   plt.show()
```

```python
# 從上面的圖檢查的結果，至少這三個欄位好像有點可疑

# AMT_INCOME_TOTAL    大部分的資料都集中在0~0.2,但是只有極少數的資料是12,12為異常值
# REGION_POPULATION_RELATIVE  只有幾筆資料在0.07 大部分的值為nan(已把nan的值drop掉)
# OBS_60_CNT_SOCIAL_CIRCLE    大部分的資料都集中在0~50,但是只有極少數的資料是350,350為異常值
```

### Hints: Emprical Cumulative Density Plot, [ECDF](https://zh.wikipedia.org/wiki/%E7%BB%8F%E9%AA%8C%E5%88%86%E5%B8%83%E5%87%BD%E6%95%B0), [ECDF with Python](https://stackoverflow.com/questions/14006520/ecdf-in-python-without-step-function)

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
dir_data = './data'
f_app = os.path.join(dir_data, 'application_train.csv')
app_train = pd.read_csv(f_app)
# print(app_train['AMT_INCOME_TOTAL'])
# 最大值離平均與中位數很遠
print(app_train['AMT_INCOME_TOTAL'].describe())
# 繪製 Empirical Cumulative Density Plot (ECDF)
def ecdf(data):
    cdfx = np.sort(data)
    cdfy = np.arange(1,len(cdfx)+1)/len(cdfx)  
    return (cdfx,cdfy)    
cdfx,cdfy= ecdf(app_train['AMT_INCOME_TOTAL'])
# print(len(cdfx),len(cdfy))
plt.xlabel('Value')
plt.ylabel('ECDF')
plt.xlim([np.nanmin(cdfx), np.nanmax(cdfx) * 1.05]) # 限制顯示圖片的範圍
plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍
plt.plot(cdfx,cdfy)
plt.show()

# 改變 y 軸的 Scale, 讓我們可以正常檢視 ECDF
plt.plot(np.log(cdfx),cdfy)
plt.xlabel('Value (log-scale)')
plt.ylabel('ECDF')

plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍

plt.show()
```

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
dir_data = './data'
f_app = os.path.join(dir_data, 'application_train.csv')
app_train = pd.read_csv(f_app)
# 最大值落在分布之外
print(app_train['OBS_60_CNT_SOCIAL_CIRCLE'].describe())
# 繪製 Empirical Cumulative Density Plot (ECDF)
def ecdf(data):
    cdfx = np.sort(data)
    cdfy = np.arange(1,len(cdfx)+1)/len(cdfx)  
    return (cdfx,cdfy)    
cdfx,cdfy= ecdf(app_train['OBS_60_CNT_SOCIAL_CIRCLE'])
plt.plot(cdfx,cdfy)
plt.xlabel('Value')
plt.ylabel('ECDF')
plt.xlim([np.nanmin(cdfx) * 0.95,np.nanmax(cdfx) * 1.05])    
plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍
plt.show()

app_train['OBS_60_CNT_SOCIAL_CIRCLE'].hist()
plt.show()
print(app_train['OBS_60_CNT_SOCIAL_CIRCLE'].value_counts().sort_index(ascending = False))
```
