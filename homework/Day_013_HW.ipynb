```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# 設定 data_path
dir_data = './data/'
```

### 之前做過的處理

```python
# 讀取資料檔
f_app_train = os.path.join(dir_data, 'application_train.csv')
app_train = pd.read_csv(f_app_train)
app_train.shape
```

```python
# 將只有兩種值的類別型欄位, 做 Label Encoder, 計算相關係數時讓這些欄位可以被包含在內
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

# 檢查每一個 column
for col in app_train:
    if app_train[col].dtype == 'object':
        # 如果只有兩種值的類別型欄位
        if len(list(app_train[col].unique())) <= 2:
            # 就做 Label Encoder, 以加入相關係數檢查
            app_train[col] = le.fit_transform(app_train[col])            
print(app_train.shape)
app_train.head()
```

```python
# 受雇日數為異常值的資料, 另外設一個欄位記錄, 並將異常的日數轉成空值 (np.nan)
app_train['DAYS_EMPLOYED_ANOM'] = app_train["DAYS_EMPLOYED"] == 365243
app_train['DAYS_EMPLOYED'].replace({365243: np.nan}, inplace = True)

# 出生日數 (DAYS_BIRTH) 取絕對值 
app_train['DAYS_BIRTH'] = abs(app_train['DAYS_BIRTH'])
```

## 練習時間
參考 Day 12 範例程式，離散化你覺得有興趣的欄位，並嘗試找出有趣的訊息

```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件
# 將只有兩種值的類別型欄位, 做 Label Encoder, 計算相關係數時讓這些欄位可以被包含在內
from sklearn.preprocessing import LabelEncoder
%matplotlib inline
plt.style.use('ggplot')
# 設定 data_path
dir_data = './data/'
# 讀取資料檔
f_app_train = os.path.join(dir_data, 'application_train.csv')
app_train = pd.read_csv(f_app_train)
le = LabelEncoder()

# 檢查每一個 column
for col in app_train:
    if app_train[col].dtype == 'object':
        # 如果只有兩種值的類別型欄位
        if len(list(app_train[col].unique())) <= 2:
            # 就做 Label Encoder, 以加入相關係數檢查
            app_train[col] = le.fit_transform(app_train[col])   
            
INCOME_DATA = app_train.loc[:, ['TARGET', 'AMT_INCOME_TOTAL']] # subset
#將NA取代為中位數
INCOME_DATA[INCOME_DATA['AMT_INCOME_TOTAL'].isnull()]['AMT_INCOME_TOTAL'] = np.median(INCOME_DATA[~INCOME_DATA['AMT_INCOME_TOTAL'].isnull()]['AMT_INCOME_TOTAL'])
#取q25 和 q75
q25 =np.percentile(INCOME_DATA['AMT_INCOME_TOTAL'], q = 25)
q75 =np.percentile(INCOME_DATA['AMT_INCOME_TOTAL'], q = 75)
bin_cut = np.linspace(q25, q75, num = 20) 
# 新增欄位 "EQUAL_WITH_AMT_INCOME_TOTAL", 對年齡做等值劃分
INCOME_DATA['EQUAL_WITH_AMT_INCOME_TOTAL'] = pd.cut(INCOME_DATA['AMT_INCOME_TOTAL'], bins = bin_cut) 
# 計算每個年齡區間的 Target、 AMT_INCOME_TOTAL 的平均值
INCOME_GROUPS = INCOME_DATA.groupby('EQUAL_WITH_AMT_INCOME_TOTAL').mean()
plt.figure(figsize = (8, 8))

# 以AMT_INCOME為區間為 x, target 為 y 繪製 barplot

px = INCOME_GROUPS.index.astype(str)
py = 100 * INCOME_GROUPS['TARGET']
sns.barplot(px, py)

# Plot labeling
plt.xticks(rotation = 75); plt.xlabel('AMT_INCOME Group'); plt.ylabel('Failure to Repay (%)')
plt.title('Failure to Repay by AMT_INCOME Group');
```
